```python
from datasets import load_dataset

dataset = load_dataset("fashion_mnist") 
```

```python
dataset['train']['image'][0].size
```

```python
labels = dataset['train']['label']
```

```python
labels = dataset['train'].features['label'].names
```

```python
clip_labels = [f"a photo of a {label}" for label in labels]
clip_labels
```

```python
label_embeddings = []
for i in clip_labels:
    text = embed_anything.embed_query([i], embeder= "Clip")
    print(text[0].embedding)
    label_embeddings.append(text[0].embedding)
```

```python
similarities = np.dot(label_embeddings, jacket[0].embedding)
max_index = np.argmax(similarities)
```

```python
print(labels[max_index])
```

```python
for i in range(len(data)):
    similarities = np.dot(label_embeddings, data[i].embedding)
    max_index = np.argmax(similarities)
    print(labels[max_index])
```
